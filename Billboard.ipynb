```python
import json
import pandas as pd
```

```python
with open('./Eminem.json') as json_data:
    Em = json.load(json_data)
```

```python
music = pd.DataFrame([song['lyrics'] for song in Em])
```

```python
music['title'] = [song['title'] for song in Em]
```

```python
music['year'] = [song['year'] for song in Em]
```

```python
music.head()
```

```python
music[music['title'] == 'The Real Slim Shady']
```

```python
music.dropna(inplace=True)
```

```python
music.reset_index(inplace=True)
```

```python
music.info()
```

```python
import regex as re
from nltk.tokenize import RegexpTokenizer
```

```python
pat = "\\[.+?\\]|\\(.+?\\)"
re_pat = re.compile(pat)
```

```python
for i in range(len(music)):
    music.loc[i,0] = re_pat.sub("", music.loc[i,0])

    
```

```python
pat = "\\n+"
re_pat = re.compile(pat)
```

```python
for i in range(len(music)):
    music.loc[i,0] = re_pat.sub(" ", music.loc[i,0])
```
