```python
def hot100(title):
    if title in billboard['Title - Modified'].values:
        return 1
    else:
        return 0
```

```python
def songsAreSame(s1, s2):    
    seqA = sm(None, s1, s2)
    seqB = sm(None, s2, s1)
    return seqA.ratio() > 0.48 or seqB.ratio() > 0.48
```

```python
songsAreSame(music.loc[326,'title'],music.loc[327,'title'])
```
